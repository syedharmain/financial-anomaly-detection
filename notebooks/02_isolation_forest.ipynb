```python
import pandas as pd

df = pd.read_csv('../data/processed/processed_data.csv')

df.head()
```

```python
from sklearn.ensemble import IsolationForest

# Remove target column (Class)
X = df.drop('Class', axis=1)

# Create model
model = IsolationForest(
    n_estimators=100,
    contamination=0.0017,
    random_state=42
)

# Train model
model.fit(X)

# Predict anomalies
df['anomaly'] = model.predict(X)

df.head()
```

```python
from sklearn.metrics import classification_report

# Convert anomaly labels (-1 → 1, 1 → 0)
df['anomaly_converted'] = df['anomaly'].apply(lambda x: 1 if x == -1 else 0)

print(classification_report(df['Class'], df['anomaly_converted']))
```

```python
model = IsolationForest(
    n_estimators=200,
    contamination=0.01,   # Increased sensitivity
    random_state=42
)

model.fit(X)
df['anomaly'] = model.predict(X)

df['anomaly_converted'] = df['anomaly'].apply(lambda x: 1 if x == -1 else 0)

from sklearn.metrics import classification_report
print(classification_report(df['Class'], df['anomaly_converted']))
```

```python
import joblib

joblib.dump(model, '../models/isolation_forest.pkl')
```
